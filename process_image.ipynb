```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
from PIL import Image
```

```python
images_path = 'data/unsplash/'
target_path = 'data/unsplash_cropped/'
```

```python
def display_rgb_image(img):
    plt.figure()
    plt.imshow(img)
    plt.show()
```

```python
def crop_image(im):
    w = im.size[0]
    h = im.size[1]

    if h > w:
        x1 = 0
        y1 = (h-w)/2
        x2 = w
        y2 = (h+w)/2

    else:
        x1 = (w-h)/2
        y1 = 0
        x2 = (w+h)/2
        y2 = h

    img = im.crop((x1, y1, x2, y2))
    return img
```

```python
folders = os.listdir(images_path)

for folder in folders:
    images = os.listdir(images_path + folder)
    for i, image in enumerate(images):
        im = Image.open(images_path + folder + '/' + image)
        c_im = crop_image(im)
        c_im = c_im.convert('RGB')
        
        directory = target_path + folder + '/'
        
        if not os.path.exists(directory):
            os.makedirs(directory)
        
        
        c_im.save(directory + image)
        
        if (i % 500 == 0):
            display_rgb_image(im)
            display_rgb_image(c_im)
            print(i)
```
