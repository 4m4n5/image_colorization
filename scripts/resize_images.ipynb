```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
from PIL import Image
```

```python
images_path = 'data/unsplash_cropped/'
target_path = 'data/unsplash_cropped_resized/'
```

```python
def display_rgb_image(img):
    plt.figure()
    plt.imshow(img)
    plt.show()
```

```python
def resize_image(im):
    img = im.resize((128, 128))
    return img
```

```python
folders = os.listdir(images_path)

for folder in folders:
    images = os.listdir(images_path + folder)
    for i, image in enumerate(images):
        im = Image.open(images_path + folder + '/' + image)
        c_im = resize_image(im)
        c_im = c_im.convert('RGB')
        
        directory = target_path + folder + '/'
        
        if not os.path.exists(directory):
            os.makedirs(directory)
        
        
        c_im.save(directory + image)
        
        if (i % 500 == 0):
            display_rgb_image(im)
            display_rgb_image(c_im)
            print(i)
```
